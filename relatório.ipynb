# Importação das bibliotecas que serão utilizadas

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

# Leitura do arquivo csv com os dados

```python
df = pd.read_csv('./data.csv')
```

# Total de horas trabalhadas

```python
total_horas = df['Horas Trabalhadas'].sum()
print(f'Total de Horas Trabalhadas: {total_horas} horas')
```

# Visualizacao com a media de horas trabalhadas ao longo da semana


```python
media_diaria_horas = df['Horas Trabalhadas'].mean()
sns.set_style('whitegrid')
plt.figure(figsize=(10, 6))
sns.lineplot(x='Dia', y='Horas Trabalhadas', data=df, marker='o', color='blue', label='Horas Trabalhadas')
sns.lineplot(x='Dia', y=media_diaria_horas, linestyle='--', color='red', label='Média de Horas Trabalhadas', data=df)
plt.xlabel('Dias da Semana')
plt.ylabel('Horas Trabalhadas')
plt.title("Media de horas trabalhadas ao longo da semena")
plt.legend()
plt.show()
```

# Total de bugs corrigidos ao longo da semana

```python
total_bugs_corrigidos = df['Bugs Corrigidos'].sum()
print(f'Total de bugs corrigidos na semana foi {total_bugs_corrigidos}')
```

# Media de bugs corrigidos ao longo da semana

```python
media_bug_corrigidos = df['Bugs Corrigidos'].mean()
plt.figure(figsize=(10,6))
sns.lineplot(x='Dia', y='Bugs Corrigidos', data=df, marker='o', color='orange', label='Bugs Corrigidos')
sns.lineplot(x='Dia', y=media_bug_corrigidos, data=df, linestyle='--', color='brown', label='Media de bugs corrigidos')
plt.xlabel('dias da semana')
plt.ylabel('Quantidade de bugs')
plt.title('Media de Bugs Corrigidos Ao Longo Da Semana')
plt.legend()
plt.grid(True)
plt.show()
```

# Total de tarefas concluidas

```python
total_tarefas = df['Tarefas Concluídas'].sum()
print(f'\n\nTotal de Tarefas Concluídas: {total_tarefas}\n')
```

# Media de tarefas concluidas ao longo da semana

```python
media_tarefas_concluidas = df['Tarefas Concluídas'].mean()
sns.lineplot(x='Dia', y='Tarefas Concluídas', data=df, marker='o', color='green', label='Tarefas Concluidas')
sns.lineplot(x='Dia', y=media_tarefas_concluidas, data=df, linestyle='--', color='purple', label='Media de tarefas concluidas')
plt.xlabel('dias da semana')
plt.title('Media de Tarefas Concluídas Ao Longo Da Semana')
plt.legend()
plt.grid(True)
plt.show()
```
